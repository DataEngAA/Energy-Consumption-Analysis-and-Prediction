##### Task 2: Handling Missing Data

```python
# Handle missing values in relevant columns
df['Temperature_C'] = df['Temperature_C'].fillna(df['Temperature_C'].mean())
df['Humidity_%'] = df['Humidity_%'].fillna(df['Humidity_%'].mean())

# For categorical columns (if applicable)
df['City'] = df['City'].fillna(df['City'].mode()[0])

# Verifying no missing values left
print(df.isnull().sum())
```

##### Task 3: Outlier Detection and Handling

```python
# Detect and handle outliers in energy consumption and temperature using boxplot
sns.boxplot(x=df['Energy_Consumption_kWh'])
plt.show()

# Capping outliers for Energy_Consumption_kWh at the 99th percentile
upper_limit = df['Energy_Consumption_kWh'].quantile(0.99)
df['Energy_Consumption_kWh'] = np.where(df['Energy_Consumption_kWh'] > upper_limit, upper_limit, df['Energy_Consumption_kWh'])

# Repeat for other numeric columns, if needed
sns.boxplot(x=df['Energy_Consumption_kWh'])
plt.show()
```

##### Task 4: Time-Series Consistency

```python
# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Check for duplicates in Date and Home_ID combination
print(df.duplicated(subset=['Date', 'Home_ID']).sum())

# Remove duplicate entries (if any)
df = df.drop_duplicates(subset=['Date', 'Home_ID'])

# Set Date as the index (if it's time-series data)
df.set_index('Date', inplace=True)
```

##### Task 5: Data Normalization

```python
from sklearn.preprocessing import StandardScaler

# Normalize numeric columns
scaler = StandardScaler()
df[['Energy_Consumption_kWh', 'Temperature_C', 'HVAC_Usage_kWh', 'Kitchen_Usage_kWh', 'Electronics_Usage_kWh']] = scaler.fit_transform(df[['Energy_Consumption_kWh', 'Temperature_C', 'HVAC_Usage_kWh', 'Kitchen_Usage_kWh', 'Electronics_Usage_kWh']])

# Check the normalization
print(df.head())
```

##### Task 4: Feature Engineering

```python
# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Ensure columns are of the correct types
df['Energy_Consumption_kWh'] = pd.to_numeric(df['Energy_Consumption_kWh'], errors='coerce')
df['Occupancy'] = pd.to_numeric(df['Occupancy'], errors='coerce')

# Check for NaN values in case of conversion issues
print("NaN values in Energy Consumption:", df['Energy_Consumption_kWh'].isna().sum())
print("NaN values in Occupancy:", df['Occupancy'].isna().sum())

# Calculate Energy per Occupant, avoiding division by zero
df['Energy_per_Occupant'] = df['Energy_Consumption_kWh'] / df['Occupancy'].replace(0, np.nan)

# Exclude non-numeric columns for the correlation heatmap
numeric_df = df.select_dtypes(include=[np.number])

# Correlation heatmap with new feature
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

## Machine Learning

 ##### Task 1: Data Splitting

```python
from sklearn.model_selection import train_test_split

# Define features X and target y
X = df[['Temperature_C', 'Humidity_%', 'HVAC_Usage_kWh', 'Kitchen_Usage_kWh', 'Electronics_Usage_kWh', 'Occupancy']]
y = df['Energy_Consumption_kWh']

# Train-test split (80/20)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)  # Respecting time-series data
```

##### Task 2: Model Selection and Training

```python
from sklearn.linear_model import LinearRegression

# Linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)
```

###### Evaluate the model using Mean Absolute Error and R-squared

```python
from sklearn.metrics import mean_absolute_error, r2_score

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Absolute Error: {mae}')
print(f'R-squared: {r2}')

# Residual plot
residuals = y_test - y_pred
sns.scatterplot(x=y_pred, y=residuals)
plt.axhline(0, color='red')
plt.show()
```

###### Analyze residuals and check for patterns or biases

```python
# Calculate residuals
residuals = y_test - y_pred

# Plot residuals vs predicted values
sns.scatterplot(x=y_pred, y=residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Predicted Energy Consumption')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()

# Residuals distribution plot
sns.histplot(residuals, kde=True)
plt.title('Residuals Distribution')
plt.show()
```

##### Task 4: Feature Importance and Interpretation
###### Identify the most important features

```python
# Coefficient analysis
coefficients = pd.DataFrame({'Feature': X_train.columns, 'Coefficient': model.coef_})
coefficients = coefficients.sort_values(by='Coefficient', ascending=False)

print(coefficients)
```

 ###### Feature importance visualization

```python
# Plot feature importance
sns.barplot(x='Coefficient', y='Feature', data=coefficients)
plt.title('Feature Importance')
plt.show()
```

##### Task 5: Predictive System and Testing
###### Build a simple predictive system

```python
# Predict on the test set
y_pred = model.predict(X_test)

# Compare predictions with actual values
comparison_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(comparison_df.head())
```

###### Visualize the predictions vs actual values

###### Scatter plot and residual plot will give you a visual comparison of the modelâ€™s performance.

```python
# Scatter plot of actual vs predicted values
plt.figure(figsize=(10,6))
sns.scatterplot(x=y_test, y=y_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # Line showing perfect predictions
plt.xlabel('Actual Energy Consumption')
plt.ylabel('Predicted Energy Consumption')
plt.title('Actual vs Predicted Energy Consumption')
plt.show()

# Residual plot (to check model errors)
sns.scatterplot(x=y_pred, y=residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Predicted Energy Consumption')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```
